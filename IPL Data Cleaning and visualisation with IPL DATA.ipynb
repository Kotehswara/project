## Encoding the data

```python
print(type(df['Team1'][0]))
```

```python
l=[]
for i in df['Team1']:
    l.append(i.strip())
df['Team1']=l
l=[]
for i in df['Team2']:
    l.append(i.strip())
df['Team2']=l #removing the extra spaces
```

```python
l1=df['Team1'].unique()
l2=df['Team2'].unique()
```

```python
en_l1={}
x=0
for i in l1:
    en_l1[i]=x
    x=x+1  
```

```python
en_l1
```

```python
df['Team1']=df['Team1'].map(en_l1)
df['Team2']=df['Team2'].map(en_l1)
```

```python
df.head(10)
```

```python
df.drop(df.columns[df.columns.str.contains('unnamed',case = False)],axis = 1, inplace = True)
```

```python
l=[]
lis=df['Toss'].unique()
for i in lis:
    l.append(i)
d_l={}
x=0
for i in l:
    s=i[0]
    if(s=='8' or s=='4' or s=='.'):
        d_l[i]=int(10000)
    else:
        d_l[i]=int(x)
        x=x+1
```

```python
df['Toss']=df['Toss'].map(d_l)
d_l
```

```python
l=df['Place'].unique()
x=0
en_pl={}
for i in l:
    en_pl[i]=x
    x=x+1
df['Place']=df['Place'].map(en_pl)
```

```python
df.head()
```

```python
lis=df['Tied'].unique()
print("Match" in lis[1])
```

```python
d={}
for i in lis:
    if(type(i)== float):
        d[i]=0
    elif ("Match abandoned" in i):
        d[i]=1
    elif("Match tied" in i):
        d[i]=2
```

```python
df['Tied']=df['Tied'].map(d)
```

```python
df['won_runs'].unique()
```

```python
lis=df['won_runs'].unique()
d={}
for i in lis:
    if(type(i)==float):
        d[i]=0
    else:
        d[i]=int(i[0:3])
df['won_runs']=df['won_runs'].map(d)
```

```python
lis=df['won_wickets'].unique()
d={}
for i in lis:
    if(type(i)==float):
        d[i]=0
    else:
        d[i]=int(i[0:3])
df['won_wickets']=df['won_wickets'].map(d)
```

```python
lis=df['Result'].unique()
l=[]
for r in lis:
    l.append(r.split(" won")[0])
s=[]
for i in df['Result']:
    s.append(i.split(" won")[0].lower())
df['Result']=s      
df.head()
```

```python
en_l={}
x=0
for i in l1:
    en_l[i.lower()]=x
    x=x+1
for i in l:
    if ("Match abandoned" in i):
        en_l[i]=20
    elif("Match tied" in i):
        en_l[i]=21
   
df['Result']=df['Result'].map(en_l)  
```
